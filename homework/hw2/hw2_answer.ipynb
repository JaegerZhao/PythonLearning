### list的使用
在Python中，list（列表）是一种有序的可变容器，可以存储不同类型的元素。列表使用方括号（[]）来表示，元素之间以逗号（,）分隔。

以下是一个示例列表的声明和初始化：

```
fruits = ['apple', 'banana', 'orange']
```

在这个例子中，我们创建了一个`fruits`列表，其中有三个字符串元素。

列表有许多有用的方法和操作，例如：

1. 访问列表元素：可以使用索引（从0开始）来访问列表中的元素。例如，`fruits[0]`将返回列表中的第一个元素。

2. 修改列表元素：列表中的元素是可变的，可以通过索引来修改它们。例如，`fruits[1] = 'grape'`将把第二个元素改为'grape'。

3. 增加元素：可以使用`append()`方法向列表末尾添加新元素。例如，`fruits.append('melon')`将在列表末尾添加一个新元素。

4. 删除元素：可以使用`remove()`方法根据元素的值来删除列表中的元素。例如，`fruits.remove('banana')`将删除列表中的'banana'元素。

5. 列表长度：可以使用`len()`函数获取列表的长度，即列表中元素的个数。

当涉及到列表操作时，有一些常见的方法和操作，下面是一些例子来说明它们：

1. 访问列表元素：
```
fruits = ['apple', 'banana', 'orange']
print(fruits[0])   # 输出: 'apple'
print(fruits[2])   # 输出: 'orange'
```

2. 修改列表元素：
```
fruits = ['apple', 'banana', 'orange']
fruits[1] = 'grape'
print(fruits)   # 输出: ['apple', 'grape', 'orange']
```

3. 添加元素到列表末尾：
```
fruits = ['apple', 'banana', 'orange']
fruits.append('melon')
print(fruits)   # 输出: ['apple', 'banana', 'orange', 'melon']
```

4. 删除列表元素：
```
fruits = ['apple', 'banana', 'orange']
fruits.remove('banana')
print(fruits)   # 输出: ['apple', 'orange']
```

5. 列表长度：
```
fruits = ['apple', 'banana', 'orange']
print(len(fruits))   # 输出: 3
```

6. 列表切片：
```
fruits = ['apple', 'banana', 'orange', 'melon', 'grape']
print(fruits[1:4])     # 输出: ['banana', 'orange', 'melon']
print(fruits[:3])      # 输出: ['apple', 'banana', 'orange']
print(fruits[2:])      # 输出: ['orange', 'melon', 'grape']
print(fruits[::2])     # 输出: ['apple', 'orange', 'grape']
```

7. 列表排序：
```
numbers = [5, 2, 8, 1, 9]
numbers.sort()
print(numbers)   # 输出: [1, 2, 5, 8, 9]
```

这些是一些常见的列表操作，可以根据具体需求使用更多的内置方法和操作来处理列表。

需要注意的一些事项：

1. 列表是可变的：与字符串不同，列表是可变的，这意味着可以修改列表中的元素，也可以添加或删除列表中的元素。

2. 列表允许重复元素：列表中可以包含重复的元素。

3. 列表可以包含不同类型的元素：Python的列表可以容纳不同类型的元素，例如整数、字符串、浮点数等。

4. 列表的索引从0开始：列表中的索引从0开始，因此第一个元素的索引是0，第二个是1，以此类推。

5. 列表是有序的：列表中的元素按照添加的顺序排列，并可以通过索引进行访问。

这些只是列表的一些基本操作和注意事项，还有很多有关列表的高级用法和方法可以进一步探索。

````python
def findKthLargest(nums, k):
  if k >= 1 and k <= len(nums) and len(nums) > 0:
    nums_ = sorted(nums,reverse = True)
    result = nums_[k-1]
    print(result)
  else:
    print('error：输入内容不符合要求')
    
nums = [3,2,1,5,6,4]
k = -1
findKthLargest(nums, k)
````

当我们编写代码时，有时候会遇到一些可能会引发异常（错误）的情况。为了避免程序崩溃或处理这些异常情况，我们可以使用try-except语句。

try-except语句的基本结构如下：

```python
try:
    # 可能引发异常的代码块
except ExceptionType:
    # 处理异常的代码块
```

代码块中的"try"部分包含我们要执行的可能会引发异常的代码。如果该代码块中的任何一行代码导致异常抛出，那么程序的执行流程将立即转到对应的"except"部分。

"except"部分是用来处理异常的地方。在这个部分，我们可以添加适当的代码来捕获和处理异常，或采取其他行动。"ExceptionType"是指我们要捕获的特定异常类型，例如ZeroDivisionError、ValueError等。如果未指定异常类型，那么"except"将捕获所有异常。

以下是一个示例，演示了如何使用try-except语句处理可能的除零错误：

```python
try:
    result = 10 / 0
except ZeroDivisionError:
    print("除零错误发生了！")
```

在这个例子中，我们尝试将10除以0，这会导致ZeroDivisionError异常。然后，异常被捕获并打印出一条消息"除零错误发生了！"。

使用try-except语句可以增加程序的健壮性，使其能够处理异常情况，并避免意外终止。此外，我们还可以在except块中执行一些特定的操作，比如记录错误日志、进行回滚或者提供备用操作。
